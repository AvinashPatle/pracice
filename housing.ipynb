```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("dataset.csv")
```

```python
print(df.shape)
df.head(5)
```

```python
import  matplotlib.pyplot as plt

plt.xlabel('area')
plt.ylabel('price')

plt.scatter(df.area,df.price,color='green',marker = '*')
```

```python
X = df.drop('price',axis= 'columns')
X
```

```python
y = df.price

y
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
x = 2000
LandAreainSqFt = [[x]]
PredictedModelResult = model.predict(LandAreainSqFt)

print(PredictedModelResult)
```

```python
m = model.coef_
print(m)
```

```python
b = model.intercept_
print(b)
```

```python
y= m*x+b

print("Price of the {0} squarfit land is:{1}".format(x,y[0]))
```
