```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
df = datasets.load_iris()
```

```python
X = df.data
y = df.target

names = df.target_names
```

```python
from sklearn.decomposition import PCA

model = PCA(n_components = 2)

y_mean = model.fit_transform(X)
```

```python
plt.figure()

color = ["red","green","orange"]

for color,i,target_names in zip(color,[0,1,2],names):
    plt.scatter(y_mean[y==i,0],y_mean[y==i,1],color = color,lw = 2,label = target_names)
    
plt.title("IRIS Clustering")
plt.show()
```
